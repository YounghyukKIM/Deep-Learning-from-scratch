```python
import matplotlib.pylab as plt
import numpy as np
```

1. 미분
    - 미분(Differentiation)은 함수의 기울기(변화량)를 구하는 방법이다.

- 1.1 나쁜 수치 미분 구현 (Bad Practice)

```python
def numerical_diff(f, x):
    h = 1e-4
    return (f(x + h) - f(x - h)) / (2 * h)
```

    - 중심 차분(central difference)을 사용하여 정확도를 높인다.
    - 전방 차분(forward difference)보다 오차가 적다.

2. 수치 미분 예제


- 2.1 함수 예시

```python
def function_1(x):
    return 0.01 * x**2 + 0.1 * x
```

```python
x = np.arange(0.0, 20.0, 0.1)
y = function_1(x)

plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x, y)
plt.show()
```

    - 0부터 20까지 x값에 대해 y값을 계산하여 그래프를 그림.



- 2.3 특정 지점에서의 미분

```python
print(numerical_diff(function_1, 5))   # 약 0.2
print(numerical_diff(function_1, 10))  # 약 0.3
```

3. 접선 구하기

- 3.1 접선 함수 구현
    - 함수 $f$에 대해 $x$점에서의 접선 방정식을 반환한다.
    - 접선 기울기 : $f′(x)$

```python
def tangent_line(f, x):
    d = numerical_diff(f, x)
    y = f(x) - d * x
    return lambda t: d * t + y
```

- 3.2 접선 그래프 그리기

```python
tf = tangent_line(function_1, 5)
y2 = tf(x)

plt.plot(x, y2)
plt.show()
```

    - 원래 곡선과 접선이 함께 표시된다.
    - 접선은 해당 점에서 원래 곡선과 같은 기울기를 가진다.

4. 편미분 (Partial Differentiation)
    - 여러 변수 중 하나에 대해서만 미분하는 것을 편미분이라고 한다.



- 4.1 다변수 함수 예시

```python
def function_2(x):
    return x[0]**2 + x[1]**2
    # 또는 return np.sum(x**2)
```

- 4.2 편미분을 위한 임시 함수
    - $x_0$ 에 대해 편미분

```python
def function_tmp1(x0):
    return x0**2 + 4.0**2
```

- $x_1$ 에 대해 편비분

```python
def function_tmp2(x1):
    return 3.0**2 + x1**2
```

- 4.3 편미분 계산

```python
print(numerical_diff(function_tmp1, 3.0))  # 약 6.0
print(numerical_diff(function_tmp2, 4.0))  # 약 8.0
```
